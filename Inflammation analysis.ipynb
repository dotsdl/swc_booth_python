Write a block of code that loops through all the files `inflammation-*` and plots the mean, max, and min across all patients for each day, as we did previously.

```python
# get filenames
filenames = glob.glob('inflammation-*.csv')
filenames.sort()

# iterate through each filename
for filename in filenames:
    print filename
    data = numpy.loadtxt(filename, delimiter=',')
    
    fig = plt.figure(figsize=(10, 3))

    ax1 = fig.add_subplot(1, 3, 1)
    ax2 = fig.add_subplot(1, 3, 2)
    ax3 = fig.add_subplot(1, 3, 3)

    # plot the means
    ax1.plot(data.mean(axis=0))
    ax1.set_ylabel("mean")
    ax1.set_ylim(top=20)

    # plot the maxes
    ax2.plot(data.max(axis=0))
    ax2.set_ylabel("max")
    ax2.set_ylim(top=20)

    # plot the minses
    ax3.plot(data.min(axis=0))
    ax3.set_ylabel("min")
    ax3.set_ylim(top=20)

    fig.tight_layout()
    plt.show()
```

Build a block of code that plots a panel of mean, max, min of each dataset, as well as the variety of weirdness the dataset exhibits. Also, print a final count of each type of weirdness observed.

```python
# get filenames
filenames = glob.glob('inflammation-*.csv')
filenames.sort()

count_susp_max = 0
count_susp_min = 0
count_okay = 0

# iterate through each filename
for filename in filenames:
    print filename
    data = numpy.loadtxt(filename, delimiter=',')
    
    if data.min(axis=0)[0] == 0 and data.max(axis=0)[20] == 20:
        print "suspcious looking maxima"
        count_susp_max += 1
    elif data.min(axis=0).sum() == 0:
        print "minima add up to zero"
        count_susp_min += 1
    else:
        print "looks okay"
        count_okay += 1
    
    fig = plt.figure(figsize=(10, 3))

    ax1 = fig.add_subplot(1, 3, 1)
    ax2 = fig.add_subplot(1, 3, 2)
    ax3 = fig.add_subplot(1, 3, 3)

    # plot the means
    ax1.plot(data.mean(axis=0))
    ax1.set_ylabel("mean")
    ax1.set_ylim(top=20)

    # plot the maxes
    ax2.plot(data.max(axis=0))
    ax2.set_ylabel("max")
    ax2.set_ylim(top=20)

    # plot the minses
    ax3.plot(data.min(axis=0))
    ax3.set_ylabel("min")
    ax3.set_ylim(top=20)

    fig.tight_layout()
    plt.show()
    
print "-------------"
print "total suspicious max: ", count_susp_max
print "total suspicious min: ", count_susp_min
print "total okay: ", count_okay
```
